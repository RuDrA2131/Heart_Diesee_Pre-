# Domain Analysis

1. **patient_id** – Unique identifier for each patient, used only for record tracking (no predictive value).

2. **heart_disease_present** – Target variable indicating whether the patient has heart disease (1 = Yes, 0 = No).

3. **slope_of_peak_exercise_st_segment** – Describes the slope of the ST segment during peak exercise, reflecting heart stress response.

4. **thal** – Thalassemia test result indicating normal, fixed defect, or reversible defect in blood flow.

5. **resting_blood_pressure** – Patient’s blood pressure measured at rest, an important cardiovascular risk indicator.

6. **chest_pain_type** – Categorizes the type of chest pain experienced, strongly associated with heart disease.

7. **num_major_vessels** – Number of major blood vessels colored by fluoroscopy, indicating vessel blockage severity.

8. **fasting_blood_sugar_gt_120_mg_per_dl** – Binary indicator of whether fasting blood sugar exceeds 120 mg/dL (diabetes risk).

9. **resting_ekg_results** – Electrocardiographic results at rest, showing heart rhythm and electrical abnormalities.

10. **serum_cholesterol_mg_per_dl** – Cholesterol level in blood, a key factor in plaque buildup and heart disease.

11. **oldpeak_eq_st_depression** – ST depression induced by exercise compared to rest, indicating myocardial ischemia.

12. **sex** – Biological sex of the patient, as heart disease risk differs between males and females.

13. **age** – Age of the patient, a major non-modifiable risk factor for heart disease.

14. **max_heart_rate_achieved** – Maximum heart rate reached during exercise, reflecting cardiovascular fitness.

15. **exercise_induced_angina** – Indicates whether chest pain occurs during exercise, signaling reduced blood flow to the heart.

## Outlier Handling

```python
num_cols = ['heart_disease_present', 'slope_of_peak_exercise_st_segment',
       'resting_blood_pressure', 'chest_pain_type', 'num_major_vessels',
       'fasting_blood_sugar_gt_120_mg_per_dl', 'resting_ekg_results',
       'serum_cholesterol_mg_per_dl', 'oldpeak_eq_st_depression', 'sex', 'age',
       'max_heart_rate_achieved', 'exercise_induced_angina']
   

for i in num_cols:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
        
    # Define Bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
        
    # Find outliers
    outliers_upper = df[df[i] > upper_bound]
    outliers_lower = df[df[i] < lower_bound]
    print(f"{i} -Upper outliers: {outliers_upper.shape}, Lower outliers: {outliers_lower.shape}")
        
    # Cap outliers using .loc
    df.loc[df[i] > upper_bound, i] = upper_bound
    df.loc[df[i] < lower_bound, i] = lower_bound
```

# Model

```python
X = df.drop("heart_disease_present",axis=1)
y = df.heart_disease_present
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

## Model Implmentation

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

log_reg = LogisticRegression()

log_reg.fit(X_train_scaled, y_train)

y_pred = log_reg.predict(X_test_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

dt = DecisionTreeClassifier()

dt.fit(X_train_scaled, y_train)

y_pred = dt.predict(X_test_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)


rf.fit(X_train_scaled, y_train)

y_pred = rf.predict(X_test_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize model
knn = KNeighborsClassifier(
    n_neighbors=5,
    weights='distance',
    metric='minkowski'
)

# Train model
knn.fit(X_train_scaled, y_train)

# Predictions
y_pred = knn.predict(X_test_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

### SVM Classifier

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize model
svc = SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale',
)

# Train model
svc.fit(X_train_scaled, y_train)

# Predictions
y_pred = svc.predict(X_test_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

 ### Model Implmentation with Hyperparameter Tuning

### Logistic Regression – Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

lr = LogisticRegression(max_iter=1000)

param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}

grid_lr = GridSearchCV(
    lr,
    param_grid_lr,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_lr.fit(X_train_scaled, y_train)

print("Best Params (LR):", grid_lr.best_params_)
print("Best Score (LR):", grid_lr.best_score_)
```

### Decision Tree – Hyperparameter Tuning

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)

param_grid_dt = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5]
}

grid_dt = GridSearchCV(
    dt,
    param_grid_dt,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_dt.fit(X_train_scaled, y_train)

print("Best Params (DT):", grid_dt.best_params_)
print("Best Score (DT):", grid_dt.best_score_)
```

### Random Forest – Hyperparameter Tuning

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_rf = GridSearchCV(
    rf,
    param_grid_rf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_rf.fit(X_train_scaled, y_train)

print("Best Params (RF):", grid_rf.best_params_)
print("Best Score (RF):", grid_rf.best_score_)
```

### KNN – Hyperparameter Tuning

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

grid_knn = GridSearchCV(
    knn,
    param_grid_knn,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_knn.fit(X_train_scaled, y_train)

print("Best Params (KNN):", grid_knn.best_params_)
print("Best Score (KNN):", grid_knn.best_score_)
```

### SVM – Hyperparameter Tuning

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svc = SVC(probability=True)

param_grid_svc = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

grid_svc = GridSearchCV(
    svc,
    param_grid_svc,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_svc.fit(X_train_scaled, y_train)
```

## ROC–AUC for ALL Models

```python
from sklearn.metrics import roc_auc_score

# Logistic Regression
y_prob_lr = grid_lr.best_estimator_.predict_proba(X_test_scaled)[:, 1]

# Decision Tree
y_prob_dt = grid_dt.best_estimator_.predict_proba(X_test_scaled)[:, 1]

# Random Forest
y_prob_rf = grid_rf.best_estimator_.predict_proba(X_test_scaled)[:, 1]

# KNN
y_prob_knn = grid_knn.best_estimator_.predict_proba(X_test_scaled)[:, 1]

# SVC 
y_prob_svc = grid_svc.best_estimator_.predict_proba(X_test_scaled)[:, 1]
```

## ROC–AUC Scores

```python
roc_scores = {
    "Logistic Regression": roc_auc_score(y_test, y_prob_lr),
    "Decision Tree": roc_auc_score(y_test, y_prob_dt),
    "Random Forest": roc_auc_score(y_test, y_prob_rf),
    "KNN": roc_auc_score(y_test, y_prob_knn),
    "SVC": roc_auc_score(y_test, y_prob_svc)
}

for model, score in roc_scores.items():
    print(f"{model}: ROC-AUC = {score:.4f}")
```

## ROC Curve Plot (All Models)

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

plt.figure(figsize=(8, 6))

for name, y_prob in zip(
    roc_scores.keys(),
    [y_prob_lr, y_prob_dt, y_prob_rf, y_prob_knn, y_prob_svc]
):
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=name)

plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves Comparison")
plt.legend()
plt.show()
```

## Final Best Model Selection

```python
results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest",
        "KNN",
        "SVC"
    ],
    "CV Accuracy": [
        grid_lr.best_score_,
        grid_dt.best_score_,
        grid_rf.best_score_,
        grid_knn.best_score_,
        grid_svc.best_score_
    ],
    "ROC-AUC": [
        roc_scores["Logistic Regression"],
        roc_scores["Decision Tree"],
        roc_scores["Random Forest"],
        roc_scores["KNN"],
        roc_scores["SVC"]
    ]
})

results.sort_values(by="ROC-AUC", ascending=False)
```
